```python
import kagglehub
path = kagglehub.dataset_download("heptapod/titanic")
```

```python
import pandas as pd
import os

# List the contents of the 'path' directory to find the correct file name.
print(os.listdir(path))

# Assuming the file is named 'titanic.csv' and is directly under 'path',
# if not, we will adjust the path based on the output of os.listdir(path).
df = pd.read_csv(os.path.join(path, 'train_and_test2.csv'))
```

```python
display(df.head(10))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=20, edgecolor='black')
plt.title('Distribution of Age using Matplotlib')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
